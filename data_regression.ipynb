```python
data1 = pd.read_csv('MICRODADOS_ENEM_2017.csv', encoding = 'latin1', sep = ';', usecols = ['NO_MUNICIPIO_RESIDENCIA',
                                                                                       'SG_UF_RESIDENCIA',
                                                                                       'NU_IDADE',
                                                                                       'TP_SEXO',
                                                                                       'TP_COR_RACA',
                                                                                       'TP_ST_CONCLUSAO',
                                                                                       'TP_ANO_CONCLUIU',
                                                                                       'TP_ESCOLA',
                                                                                       'TP_DEPENDENCIA_ADM_ESC',
                                                                                       'NU_NOTA_CH',
                                                                                       'NU_NOTA_CN',
                                                                                       'NU_NOTA_LC',
                                                                                       'NU_NOTA_LC',
                                                                                       'NU_NOTA_MT',
                                                                                       'NU_NOTA_REDACAO',
                                                                                       'Q001',
                                                                                       'Q002',
                                                                                       'Q006',
                                                                                       'Q025'])
```

```python
data1.head()
```

```python
data1.shape
```

```python
data1.dropna(how = 'any', inplace = True, axis = 0)
```

```python
data = data1.copy()
```

```python
data.shape
```

```python
data.info()
```

```python
data.head()
```

```python
data.drop('NO_MUNICIPIO_RESIDENCIA', axis = 1, inplace = True)
data['NU_NOTA_MED'] = (data.NU_NOTA_CH + data.NU_NOTA_CN + data.NU_NOTA_LC + data.NU_NOTA_MT + data.NU_NOTA_REDACAO)/5
```

```python
data.head()
```

```python
data.drop(['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO'], axis = 1, inplace = True)
```

```python
data.drop(['Q002','Q001','TP_DEPENDENCIA_ADM_ESC','TP_ST_CONCLUSAO'], axis = 1 , inplace = True)
```

```python
data.head()
```

```python
raca = {0: 'Nao_declarado', 1:'Branca', 2:'Preta', 3:'Parda', 4:'Amarela', 5:'Indigena'}
escola = {1: 'Nao_respondeu', 2:'Publica', 3: 'Privado', 4:'Exterior'}
q006 = {'A':0, 'B':937, 'C':1405, 'D':1874, 'E':2342, 'F':2811, 'G':3748, 'H':4685,
       'I':5622, 'J':6559, 'K':7496, 'L':8433, 'M':9370, 'N':1244, 'O':14055, 'P':18740, 'Q':22000}
q025 = {'A':'nao', 'B':'sim'}
```

```python
data.TP_COR_RACA = data.TP_COR_RACA.map(raca)
data.TP_ESCOLA = data.TP_ESCOLA.map(escola)
data.Q006 = data.Q006.map(q006)
data.Q025 = data.Q025.map(q025)
data.head()
```

```python
#Os dados foram categorizados em colunas com o valor 0 ou 1
```

```python
data = pd.get_dummies(data, columns=['SG_UF_RESIDENCIA','TP_SEXO','TP_COR_RACA','TP_ESCOLA','Q006','Q025'])
```

```python
data.head()
```

```python
data.columns
```

```python
Y = data.NU_NOTA_MED
data.drop('NU_NOTA_MED', axis = 1, inplace = True)
```

```python
#Fazendo a normalizacao das variaveis independentes e da variavel dependente
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(data)
```

```python
data = scaler.transform(data)
```

```python
scaler.fit(np.array(Y).reshape(-1,1))
Y = scaler.transform(np.array(Y).reshape(-1,1))
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, Y_train)
lm.score(X_train, Y_train)
```
