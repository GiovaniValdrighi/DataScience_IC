```python
#leitura dos dados
data1 = pd.read_csv('MICRODADOS_ENEM_2017.csv', encoding = 'latin1', sep = ';', usecols = ['SG_UF_RESIDENCIA',
                                                                                       'NU_IDADE',
                                                                                       'TP_SEXO',
                                                                                       'TP_COR_RACA',
                                                                                       'TP_ST_CONCLUSAO',
                                                                                       'TP_ANO_CONCLUIU',
                                                                                       'TP_ESCOLA',
                                                                                       'TP_DEPENDENCIA_ADM_ESC',
                                                                                       'NU_NOTA_CH',
                                                                                       'NU_NOTA_CN',
                                                                                       'NU_NOTA_LC',
                                                                                       'NU_NOTA_LC',
                                                                                       'NU_NOTA_MT',
                                                                                       'NU_NOTA_REDACAO',
                                                                                       'Q001',
                                                                                       'Q002',
                                                                                       'Q006',
                                                                                       'Q025'])
#limpeza dos campos NaN
data1.dropna(how = 'any', inplace = True, axis = 0)
data = data1.copy()
#criação do target
data['NU_NOTA_MED'] = (data.NU_NOTA_CH + data.NU_NOTA_CN + data.NU_NOTA_LC + data.NU_NOTA_MT + data.NU_NOTA_REDACAO)/5
#apgando algumas colunas que não quero
data.drop(['NO_MUNICIPIO_RESIDENCIA',
           'Q002',
           'Q001',
           'TP_DEPENDENCIA_ADM_ESC',
           'TP_ST_CONCLUSAO',
           'NU_NOTA_CN',
           'NU_NOTA_CH',
           'NU_NOTA_LC',
           'NU_NOTA_MT',
           'NU_NOTA_REDACAO'], axis = 1, inplace = True)
```

```python
#criação dos dicionários para mapeamento
raca = {0: 'nao_declarado', 1:'branca', 2:'preta', 3:'parda', 4:'amarela', 5:'indigena'}
escola = {1: 'nao_respondeu', 2:'publica', 3: 'privado', 4:'exterior'}
q006 = {'A':0, 'B':900, 'C':1400, 'D':1900, 'E':2300, 'F':2800, 'G':3700, 'H':4600,
       'I':5600, 'J':6600, 'K':7500, 'L':8400, 'M':9400, 'N':12000, 'O':14000, 'P':18700, 'Q':22000}
q025 = {'A':'nao', 'B':'sim'}
#fazendo o mapeamento
data.TP_COR_RACA = data.TP_COR_RACA.map(raca)
data.TP_ESCOLA = data.TP_ESCOLA.map(escola)
data.Q006 = data.Q006.map(q006)
data.Q025 = data.Q025.map(q025)
#categorizando os dados (todos serão colunas indivuduais com valores 0 ou 1)
data = pd.get_dummies(data, columns=['SG_UF_RESIDENCIA','TP_SEXO','TP_COR_RACA','TP_ESCOLA','Q006','Q025'])
```

```python
#criando o target e tirando o target do dataframe
target = data.NU_NOTA_MED
data.drop('NU_NOTA_MED', axis = 1, inplace = True)
#Fazendo a normalizacao das variaveis independentes e da variavel dependente
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(data)
data = scaler.transform(data)
scaler = StandardScaler()
scaler.fit(np.array(target).reshape(-1,1))
Y = scaler.transform(np.array(target).reshape(-1,1))
```

```python
#rodando o PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95)
pca.fit(data)
data = pca.transform(data)
```

```python
#criando grupos de treino e teste
from sklearn.model_selection import train_test_split
X_train, X_teste, Y_train, Y_test = train_test_split(data, target , test_size = 0.25, random_state = 0)
```

```python
#rodando o modelo linear
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, Y_train)
lm.score(X_train, Y_train)
```

```python
lm.score(X_teste,Y_test)
```

## Mais uma tentativa, dessa vez usando cross-validation

```python
#leitura dos dados
data1 = pd.read_csv('MICRODADOS_ENEM_2017.csv', encoding = 'latin1', sep = ';', usecols = ['SG_UF_RESIDENCIA',
                                                                                       'NU_IDADE',
                                                                                       'TP_SEXO',
                                                                                       'TP_COR_RACA',
                                                                                       'TP_ST_CONCLUSAO',
                                                                                       'TP_ANO_CONCLUIU',
                                                                                       'TP_ESCOLA',
                                                                                       'TP_DEPENDENCIA_ADM_ESC',
                                                                                       'NU_NOTA_CH',
                                                                                       'NU_NOTA_CN',
                                                                                       'NU_NOTA_LC',
                                                                                       'NU_NOTA_LC',
                                                                                       'NU_NOTA_MT',
                                                                                       'NU_NOTA_REDACAO',
                                                                                       'Q001',
                                                                                       'Q002',
                                                                                       'Q006',
                                                                                       'Q025'])
#limpeza dos campos NaN
data1.dropna(how = 'any', inplace = True, axis = 0)
data = data1.copy()
#criação do target
data['NU_NOTA_MED'] = (data.NU_NOTA_CH + data.NU_NOTA_CN + data.NU_NOTA_LC + data.NU_NOTA_MT + data.NU_NOTA_REDACAO)/5
#apgando colunas das notas
data.drop(['NU_NOTA_CN',
           'NU_NOTA_CH',
           'NU_NOTA_LC',
           'NU_NOTA_MT',
           'NU_NOTA_REDACAO'], axis = 1, inplace = True)
#criando as dummies
data = pd.get_dummies(data, columns=['SG_UF_RESIDENCIA','TP_SEXO','TP_COR_RACA','TP_ST_CONCLUSAO','TP_ESCOLA','TP_DEPENDENCIA_ADM_ESC','Q001','Q002','Q006','Q025'])
#criando o target e tirando o target do dataframe
target = data.NU_NOTA_MED
data.drop('NU_NOTA_MED', axis = 1, inplace = True)
```

```python
data.head()
```

```python
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
def modelo_linear():
    kf = KFold(n_splits = 5)
    coe = []
    scor = []
    for train_index, test_index in kf.split(data):
        #separando em x,y treino e x,y teste
        x_train, x_test = np.array(data)[train_index], np.array(data)[test_index]
        y_train, y_test = np.array(target)[train_index], np.array(target)[test_index]
        #fit o scaler com base no train, transform train e test
        scaler = StandardScaler()
        scaler.fit(x_train)
        x_train_scaled = scaler.transform(x_train)
        x_test_scaled = scaler.transform(x_test)
        #fit o pca com base no train, transform train e test
        pca = PCA(0.95)
        pca.fit(x_train_scaled)
        x_train_scaled = pca.transform(x_train_scaled)
        x_test = pca.transform(x_test_scaled)
        #fitando o linear regression no train, avaliando no test
        lm = LinearRegression()
        lm.fit(x_train_scaled, y_train)        
        scor.append(lm.score(x_test, y_test))
        coe.append(lm.coef_)
    return scor, coe
```
