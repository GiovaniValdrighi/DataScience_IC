### First Idea: Essay Note vs Math Note
My first idea is to see if there is some relation between the two notes, essay and math. I'll be using a scatter plot, the x axis will the math note and the y axis will be the essay note.

```python
data_s.NU_NOTA_REDACAO.isnull().sum() #checking if there is any NaN value
```

```python
#import plotly
#plotly.tools.set_credentials_file(username='giovanivaldrighi', api_key='k5VNr4JOaRKvVdy63v4l')
```

```python
#getting the two columns
py.offline.init_notebook_mode(connected=True)
math_x = np.array(data_s.NU_NOTA_MT)
essay_y =data_s.NU_NOTA_REDACAO
trace = go.Scatter(x=math_x,y=essay_y,mode = 'markers')
layout = go.Layout(title = 'Math x Essay', xaxis=dict(title = 'Math notes'), yaxis=dict(title = 'Essay notes'))
data1 = [trace]
fig = go.Figure(data = data1, layout=layout)
py.offline.iplot(fig, filename = 'first scatter')
```

### Second idea: Public means x Private means
I know that private schools have better results, but I want to be able to see how much of it.

```python
data_s['Mean_notes'] = (data_s.NU_NOTA_CH+data_s.NU_NOTA_CN+data_s.NU_NOTA_LC + data_s.NU_NOTA_MT + data_s.NU_NOTA_REDACAO)/5
data_s.Mean_notes.head()
```

```python
data_public = go.Histogram(x = np.array(data_s.Mean_notes[data_s.TP_ESCOLA == 1]), name= 'Public')
data_private = go.Histogram(x =np.array(data_s.Mean_notes[data_s.TP_ESCOLA == 2]), name= 'Private')
#layout = go.Layout(title = 'Math x Essay', xaxis=dict(title = 'Math notes'), yaxis=dict(title = 'Essay notes'))
#fig = go.Figure(data = data1, layout=layout)
data2 = [data_public, data_private]
py.offline.iplot(data2, filename = 'first histogram')
```

Interpreting the graphic, it looks like that public schools have best results than private schools, because the big "mass" of the blue histogram is more to right (bigger values of x) than the big "mass" of the orange histogram. Because that I'll take a look at the mean note of public and schools.
