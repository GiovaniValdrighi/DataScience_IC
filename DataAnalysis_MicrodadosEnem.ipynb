# Data Analysis of ENEM_SP data

```python
import pandas as pd
import numpy as np
import plotly as py
import plotly.graph_objs as go
```

```python
#opening the dataframe
data = pd.read_csv("dadosSP.csv")
data.head()
```

```python
data = data.drop('Unnamed: 0', axis = 1) #I don't what happened to this column, but it was student ID
```

I tried to plot using all my data from SP but got an alert from pyplot recomend me to plot just data with less than 40k points. So I'll get random 40k rows from the data.

```python
data_s = data.sample( n = 10000)
data_s.head()
```

## First Idea: Essay Note vs Math Note
My first idea is to see if there is some relation between the two notes, essay and math. I'll be using a scatter plot, the x axis will the math note and the y axis will be the essay note.

```python
data_s.NU_NOTA_REDACAO.isnull().sum() #checking if there is any NaN value
```

```python
#import plotly
#plotly.tools.set_credentials_file(username='giovanivaldrighi', api_key='k5VNr4JOaRKvVdy63v4l')
```

```python
#getting the two columns
py.offline.init_notebook_mode(connected=True)
math_x = np.array(data_s.NU_NOTA_MT)
essay_y =data_s.NU_NOTA_REDACAO
trace = go.Scatter(x=math_x,y=essay_y,mode = 'markers')
layout = go.Layout(title = 'Math x Essay', xaxis=dict(title = 'Math notes'), yaxis=dict(title = 'Essay notes'))
data1 = [trace]
fig = go.Figure(data = data1, layout=layout)
py.offline.iplot(fig, filename = 'first scatter')
```

## Second idea: Public means x Private means
I know that private schools have better results, but I want to be able to see how much of it.

```python
data_s['Mean_notes'] = (data_s.NU_NOTA_CH+data_s.NU_NOTA_CN+data_s.NU_NOTA_LC + data_s.NU_NOTA_MT + data_s.NU_NOTA_REDACAO)/5
data_s.Mean_notes.head()
```

```python
data_public = go.Histogram(x = np.array(data_s.Mean_notes[data_s.TP_ESCOLA == 2]), name= 'Public')
data_private = go.Histogram(x =np.array(data_s.Mean_notes[data_s.TP_ESCOLA == 3]), name= 'Private')
data2 = [data_public, data_private]
layout = go.Layout(title = 'Public and Private schools', xaxis=dict(title = 'Mean note'), yaxis=dict(title = 'Probability'))
fig = go.Figure(data = data2, layout=layout)
py.offline.iplot(fig, filename = 'first histogram')
```

Looking at both histograms, we can see that they appear to work like a normal distribution. But we have that the public histogram is more focused on smaller results (around 500) and the private histogram is more focused on bigger results (around 600), but also, the notes in the private histogram is better distributed. 
But my histogram didn't show the data correctly, because of the different number of students in private and public schools, the histograms have different "height".

```python
data_public = go.Histogram(x = np.array(data_s.Mean_notes[data_s.TP_ESCOLA == 2]), name= 'Public', histnorm='probability')
data_private = go.Histogram(x =np.array(data_s.Mean_notes[data_s.TP_ESCOLA == 3]), name= 'Private', histnorm='probability')
data2 = [data_public, data_private]
layout = go.Layout(title = 'Public and Private schools', xaxis=dict(title = 'Mean note'), yaxis=dict(title = 'Probability'))
fig = go.Figure(data = data2, layout=layout)
py.offline.iplot(fig, filename = 'second histogram')
```

## Third idea: years since college conclusion
I want to see what happen with the mean notes accordly to the number of years since college conclusion. What is expected is the mean to declite.

```python
data_years = go.Scatter(y = data_s.groupby('TP_ANO_CONCLUIU').mean().Mean_notes)
data3 = [data_years]
layout = go.Layout(title = 'Results grouped by year since graduation', xaxis=dict(title = 'Years'), yaxis=dict(title = 'Mean notes'))
fig = go.Figure(data = data3, layout=layout)
py.offline.iplot(fig, filename = 'first line chart')
```

Was we can see, in the first year after conclusion, the notes are bigger, probably because most of students focus one more year on ENEM, then the notes decline because most of them give up on the first year.
I want to see the results separeted by private and public schools.

```python
data_years_undefined = go.Scatter(y = data_s[data_s.TP_ESCOLA == 1].groupby('TP_ANO_CONCLUIU').mean().Mean_notes, name = 'Not determined')
data_years_public = go.Scatter(y = data_s[data_s.TP_ESCOLA == 2].groupby('TP_ANO_CONCLUIU').mean().Mean_notes, name = 'Public')
data_years_private = go.Scatter(y = data_s[data_s.TP_ESCOLA == 3].groupby('TP_ANO_CONCLUIU').mean().Mean_notes, name = 'Private')
data4 = [data_years_undefined, data_years_public, data_years_private]
layout = go.Layout(title = 'Results grouped by years since graduation - Public x Private x Not determined', xaxis=dict(title = 'Years'), yaxis=dict(title = 'Mean notes'))
fig = go.Figure(data = data4, layout=layout)
py.offline.iplot(fig, filename = 'second line chart')
```

Here I found a problem on the data, the applicants that gave the information about how many years have passed since they finished college doesn't gave any information about their old schools, if it was public or private.
Since I don't have the information about years since graduation, I'll work with the age.

```python
x_age = list(range(14,74))
data_age_undefined = go.Scatter(x =x_age,y = data_s[data_s.TP_ESCOLA == 1].groupby('NU_IDADE').mean().Mean_notes, name = 'Not determined')
data_age_public = go.Scatter(x =x_age, y = data_s[data_s.TP_ESCOLA == 2].groupby('NU_IDADE').mean().Mean_notes, name = 'Public')
data_age_private = go.Scatter(x =x_age, y = data_s[data_s.TP_ESCOLA == 3].groupby('NU_IDADE').mean().Mean_notes, name = 'Private')
data5 = [data_age_undefined, data_age_public, data_age_private]
layout = go.Layout(title = 'Results grouped by age - Public x Private x Not determined', xaxis=dict(title = 'Age'), yaxis=dict(title = 'Mean notes'))
fig = go.Figure(data = data5, layout=layout)
py.offline.iplot(fig, filename = 'third line chart')
```
