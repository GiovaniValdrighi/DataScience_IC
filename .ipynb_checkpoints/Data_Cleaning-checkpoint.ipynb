## MICRO DADOS ENEM 2017

I'm using the data of ENEM 2017 from http://inep.gov.br/microdados.
In this notebook I'll open, get only the data from State SP, throw away the columns and rows that I don't need and save it as one new csv. 

```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv("MICRODADOS_ENEM_2017.csv", encoding = "latin1", sep = ";")
dados.head()
```

```python
dados.info()
```

```python
cols = dados.columns.values.tolist()
print(len(cols))
```

```python
for elem in cols:
    print(elem + ";")
    print()
```

Reading a .xlsx in the paste of data, I could find what each column is. We have 137 columns, the first ones are description of the applicant (like age, sex...), then We have a lot of columns about any special need required. Then we have information about presence or abscense in each test, the code of each teste, the note of each test, vector with the answers for each test, information about the essay test and the note of the essay test. Then We have 27 questions about social and economics conditions of the applicant.
We will use some of the descriptions of the applicant, the information about presence (we gonna look only to the ones that were in every test) and the notes of the tests.

```python
cols = ['NO_MUNICIPIO_RESIDENCIA','NU_IDADE', 'TP_SEXO','TP_COR_RACA','TP_ST_CONCLUSAO','TP_ANO_CONCLUIU','TP_ESCOLA','IN_TREINEIRO','TP_DEPENDENCIA_ADM_ESC', 'TP_PRESENCA_CN','TP_PRESENCA_CH','TP_PRESENCA_LC','TP_PRESENCA_MT','TP_STATUS_REDACAO','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO']
```

```python
dados_SP = dados[dados.SG_UF_RESIDENCIA == "SP"]
dados_SP = dados_SP[cols]
dados_SP.head()
```

Getting only the data of the applicants that were present in every test.

```python
dados_SP = dados_SP[(dados_SP.TP_PRESENCA_CH == 1) & (dados_SP.TP_PRESENCA_CN == 1) & (dados_SP.TP_PRESENCA_LC == 1) & (dados_SP.TP_PRESENCA_MT == 1) & (dados_SP.TP_STATUS_REDACAO == 1)]
dados_SP.head()
```

Deleting the columns of the status of presence:

```python
cols = ['NO_MUNICIPIO_RESIDENCIA','NU_IDADE', 'TP_SEXO','TP_COR_RACA','TP_ST_CONCLUSAO','TP_ANO_CONCLUIU','TP_ESCOLA','IN_TREINEIRO','TP_DEPENDENCIA_ADM_ESC', 'NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO']
```

```python
dados_SP = dados_SP[cols]
```
